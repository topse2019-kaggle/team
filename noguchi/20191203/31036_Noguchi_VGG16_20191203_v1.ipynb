画像をImageNetから直接持ってきたいが、使い方分からず。

```
img1 = "/content/drive/My Drive/TopSE/ソフトウェア開発実践演習/ImageNet_Input/train/cat.11679.jpg"
img2 = "/content/drive/My Drive/TopSE/ソフトウェア開発実践演習/ImageNet_Input/train/dog.2811.jpg"
img3 = "/content/drive/My Drive/TopSE/ソフトウェア開発実践演習/ImageNet_Input/train/cat.1.jpg"
img4 = "/content/drive/My Drive/TopSE/ソフトウェア開発実践演習/ImageNet_Input/train/dog.1.jpg"
imgs = [img1, img2, img3, img4]

def _load_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)
    return img 

def _get_predictions(_model):
    f, ax = plt.subplots(1, 4)
    f.set_size_inches(20, 10) #(80, 40)
    for i in range(4):
        ax[i].imshow(Image.open(imgs[i]).resize((224, 224), Image.ANTIALIAS)) #(200, 200)
    plt.show()
    
    f, axes = plt.subplots(1, 4)
    f.set_size_inches(15, 2) #(80, 20)
    for i,img_path in enumerate(imgs):
        img = _load_image(img_path)
        preds = decode_predictions(_model.predict(img), top=3)[0]
        b = sns.barplot(y=[c[1] for c in preds], x=[c[2] for c in preds], color="gray", ax=axes[i])
        b.tick_params(labelsize=10) #(labelsize=55)
        f.tight_layout()
```

```
# KerasのAPIの学習済みVGG16を用いて4枚の画像を判定。
from keras.applications.vgg16 import VGG16
#vgg16_weights = '../input/vgg16/vgg16_weights_tf_dim_ordering_tf_kernels.h5'
vgg16_weights = '/content/drive/My Drive/TopSE/ソフトウェア開発実践演習/ImageNet_Input/vgg16_weights_tf_dim_ordering_tf_kernels.h5'
vgg16_model = VGG16(weights=vgg16_weights)
_get_predictions(vgg16_model)
```
