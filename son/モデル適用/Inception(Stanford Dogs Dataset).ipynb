```
import glob
import shutil

foruda = os.listdir(path='.')
path = os.getcwd()  

for a in foruda:
 if a.find("-") != -1:
  files = os.listdir(path + "/" + a)  
  count = len(files)  
  files = glob.glob("./"+a +"/*")
  i = 0
  for file in files:
    i = 1 + i
    if i < count*1/4:
     data_dir = "./val/" + a
     if not os.path.exists(data_dir):
      os.mkdir(data_dir)
     shutil.move(file, './val/'+ a + "/")
    else:
     data_dir = "./train/" + a
     if not os.path.exists(data_dir):
      os.mkdir(data_dir)
     shutil.move(file, './train/'+ a + "/")
  shutil.rmtree("./"+a)
```

```
from lib.lib import network_finetuning as network
from lib.lib import architecture
from lib.lib import dataloader as dl
from lib.lib import optimizer
from lib.lib import trainer

import torchvision
```

```
%%time
net = network.InceptionV3(class_size=120)
```

```
%%time
batch_size = 128
train_dataset = torchvision.datasets.ImageFolder(root='./data/images/Images/train', transform=dl.simple_transform(resize=299))
test_dataset = torchvision.datasets.ImageFolder(root='./data/images/Images/val', transform=dl.simple_transform(resize=299))
data_loader = dl.DataLoader(train_dataset, test_dataset, batch_size=batch_size)
```

```
%%time
model = architecture.CNN_Architecture(net)
```

```
%%time
model.train(data_loader, epoch_count=50, is_inception=True)
```

```
%%time
model.predict(data_loader['val'])
```
