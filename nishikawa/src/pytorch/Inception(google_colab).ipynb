```
from lib import network_finetuning as network
from lib import architecture
from lib import dataloader as dl
from lib import optimizer
from lib import trainer

import torchvision
```

```
%%time
net = network.InceptionV3(class_size=2)
```

```
%%time
batch_size = 128
train_dataset = torchvision.datasets.ImageFolder(root='./data/cell_images/train', transform=dl.simple_transform(resize=299))
test_dataset = torchvision.datasets.ImageFolder(root='./data/cell_images/test', transform=dl.simple_transform(resize=299))
data_loader = dl.DataLoader(train_dataset, test_dataset, batch_size=batch_size)
```

```
%%time
model = architecture.CNN_Architecture(net)
```

```
%%time
model.train(data_loader, epoch_count=50, is_inception=True)
```
